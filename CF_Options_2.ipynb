### 蝶式成本

```python

# 1. 筛选 Call 合约并计算隐含方差
df_calls = df[df['is_call']].copy()
df_calls['implied_var'] = df_calls['iv']**2 * df_calls['T']

# 2. 生成蝶式套利信号：cost0 < 0
def detect_butterfly(group):
    g = group.sort_values('strike').reset_index(drop=True)
    signals = []
    for i in range(len(g) - 2):
        v0, v1, v2 = g.loc[i, 'implied_var'], g.loc[i+1, 'implied_var'], g.loc[i+2, 'implied_var']
        if (v0 - 2*v1 + v2) < 0:
            k0, k1, k2 = g.loc[i, 'strike'], g.loc[i+1, 'strike'], g.loc[i+2, 'strike']
            c0, c1, c2 = g.loc[i, 'twap'], g.loc[i+1, 'twap'], g.loc[i+2, 'twap']
            cost0 = c0 - 2*c1 + c2
            if cost0 < 0:
                signals.append({
                    'minute': g.loc[i+1, 'minute'],
                    'T':       g.loc[i+1, 'T'],
                    'k0':      k0,
                    'k1':      k1,
                    'k2':      k2,
                    'cost0':   cost0
                })
    return pd.DataFrame(signals)

arb_signals = (
    df_calls
    .groupby(['minute', 'T'], group_keys=False)
    .apply(detect_butterfly)
    .reset_index(drop=True)
)


print(arb_signals.head())
```

### 回测

```python

# 3. 构建 TWAP 价格映射
price_table = df_calls.set_index(['minute','T','strike'])['twap']

# 4. 回测：入场后等待 cost >= 0 再平仓
trades = []
for sig in arb_signals.itertuples(index=False):
    t0, T0, k0, k1, k2, cost0 = sig.minute, sig.T, sig.k0, sig.k1, sig.k2, sig.cost0
    # 遍历后续所有时间点
    times = sorted(df_calls[df_calls['T'] == T0]['minute'].unique())
    for t1 in times:
        if t1 <= t0:
            continue
        try:
            c0 = price_table.loc[(t1, T0, k0)]
            c1 = price_table.loc[(t1, T0, k1)]
            c2 = price_table.loc[(t1, T0, k2)]
        except KeyError:
            continue
        cost1 = c0 - 2*c1 + c2
        if cost1 >= 0:
            trades.append({
                't0':         t0,
                't1':         t1,
                'k0':         k0,
                'k1':         k1,
                'k2':         k2,
                'cost_entry': cost0,
                'cost_exit':  cost1,
                'pnl':        cost1 - cost0
            })
            break

# 5. 结果统计
trades_df = pd.DataFrame(trades)
print("Trades:", len(trades_df))
print("Total PnL:", trades_df['pnl'].sum())
print("Avg PnL per trade:", trades_df['pnl'].mean())
print("Win rate:", (trades_df['pnl'] > 0).mean())
```
